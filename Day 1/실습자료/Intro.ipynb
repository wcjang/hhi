# Intro

### Python의 소개

#### 1. python의 소개

- 간결한 문법으로 입문자가 이해하기 쉽고, 다양한 분야에 활용가능하다
- 컴퓨터 교육 프로그래밍 뿐만 아니라 기업의 실무를 위해서도 많이 사용되는 언어로 활용범위가 점점 증가하고 있다
- 무료로 설치 가능하며, 통계 분석 뿐만 아니라 머신러닝 및 딥러닝 분야에서 특히 많이 사용된다.

#### 2. Python 설치하기

- Anaconda는 데이터 분석을 위해 사용되는 R과 파이썬 등을 쉽게 사용가능하도록 다양한 오픈 소스를 모아놓은 데이터 과학 플랫폼이다.
- 공식홈페이지에서도 파이썬을 설치할 수 있으나, Anaconda를 이용하면 손쉽게 파이썬의 모든 데이터 과학 패키지를 설치, 관리할 수 있다.

- https://www.anaconda.com/  에 접속 후 다운로드 버튼 클릭한다
