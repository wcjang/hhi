```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import statsmodels.stats.api as sm
import seaborn as sns
```

# 3번

```python
from scipy import stats
```

$B(400, (3/4)^{10})$

```python
p = (3/4)**10

np.random.seed(20)
x = np.arange(1,47)

fig, ax = plt.subplots(figsize=(11,3.5))
rv = stats.binom.pmf(x, n=400, p=p)
sns.barplot(x=x, y=rv, ax =ax)
plt.show()
```

```python
value = stats.binom.cdf(1, n=400, p = p)
print(f'찾고자 하는 확률은 {value}입니다.')
```

# 4번

```python
np.random.seed(11)
sample = stats.uniform.rvs(0,100, size=300)

my_samples = []
for _ in range(10000):
    x = np.random.choice(sample, size=300, replace=True)
    my_samples.append(np.median(x))
```

```python
plt.hist(sample) # 기존 sample
plt.show()
```

```python
np.median(sample)
```

```python
plt.hist(my_samples)
plt.show()
```

```python
med = np.median(sample)
confidence_interval = list(pd.Series(my_samples).quantile([0.025, 0.975]))
print(med)
print(confidence_interval)
```

```python
plt.hist(my_samples)
plt.axvline(x=np.median(sample), color= 'red')
plt.plot(confidence_interval, [2000,2000], color='black')
for x in confidence_interval:
    plt.plot([x, x], [1800, 2200], color='black')
plt.show()
```
